```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# Identify non-numeric columns in xtrain
non_numeric_cols = xtrain.select_dtypes(exclude=['number']).columns

# Handle non-numeric columns (example: drop them)
xtrain_numeric = xtrain.drop(columns=non_numeric_cols)
xtest_numeric = xtest.drop(columns=non_numeric_cols)  # Apply the same transformation to xtest

imputer = SimpleImputer(strategy='mean')

xtrain_imputed = imputer.fit_transform(xtrain_numeric)
xtest_imputed = imputer.transform(xtest_numeric)

LOR = LogisticRegression()
LOR.fit(xtrain_imputed, ytrain)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# Use xtest_numeric which has the same columns as the data used during fit
xtest_imputed = imputer.transform(xtest_numeric)

pred = LOR.predict(xtest_imputed)
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(ytest,pred)
```
